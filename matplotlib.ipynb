```python
import matplotlib.pyplot as plt
import numpy as np

x1=np.array([1,2,3,4,5])
y1=x1*x1+2
plt.subplot(2,3,1)
plt.plot(x1,y1,'g--')

x2=x1*5
y2=y1*5
plt.subplot(2,3,2)
plt.plot(x2,y2,'r--')

x3=x2*x1
y3=y2*y1
plt.subplot(2,3,4)
plt.plot(x3,y3,'y--')
plt.show()
```

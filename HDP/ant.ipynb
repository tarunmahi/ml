```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib as plt
```

```python
ant_df=pd.read_csv('./data/PROMISE/ant-1.3.csv')
```

```python
sns.set(rc={'figure.figsize':(10,5)})
sns.swarmplot(x=ant_df['loc'],y=ant_df['bug'])
```

```python
sns.distplot(ant_df['loc'],kde=False,color='red')
```

```python
sns.set(rc={'figure.figsize':(5,5)})
sns.kdeplot(ant_df['loc'],shade=True)
```
